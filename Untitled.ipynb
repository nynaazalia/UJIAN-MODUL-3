# C. Model Building & Hyper-parameter Tuning (30 poin)

```python
from sklearn.model_selection import train_test_split,GridSearchCV,cross_val_score,KFold
from sklearn.preprocessing import StandardScaler  
from sklearn.metrics import accuracy_score
from sklearn.neighbors import KNeighborsClassifier
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
```

I'll try some classification models

```python
X = df1.drop(['is_canceled'], axis=1)
Y = df1['is_canceled']
# SPLITTING TRAIN AND TEST DATA 90
x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.10, shuffle=False)
# I use 90 percent train data because based on my experience. More train data, better model performance
```

```python
models = [LogisticRegression(),KNeighborsClassifier(),DecisionTreeClassifier(),RandomForestClassifier()]
names = ["LogisticRegression","KNN","DecisionTree","Random_Forest"]
```

```python
print('Default model validation accuracies for the train data:', end = "\n\n")
for name, model in zip(names, models):
    model.fit(x_train, y_train)
    y_pred = model.predict(x_test) 
    print(name,':',"%.3f" % accuracy_score(y_pred, y_test))
```

```python
# Wow, high accuracy in 3 models. We have to check to make sure it's not an overfitting models
# KNN has lower accuracy so I decided to not use that
```

## KFolds

```python
from sklearn.model_selection import KFold

kfold = KFold(n_splits =10 , shuffle=True, random_state=0)
clf = LogisticRegression()
scoring = 'accuracy'
results = cross_val_score(clf, x_train, y_train, cv= kfold, scoring=scoring)
print(results)
print("10 Fold cross validation average accuracy: %.3f" %(results.mean()))
```

```python
kfold = KFold(n_splits =10 , shuffle=True, random_state=0)
clf = DecisionTreeClassifier()
scoring = 'accuracy'
results = cross_val_score(clf, x_train, y_train, cv= kfold, scoring=scoring)
print(results)
print("10 Fold cross validation average accuracy: %.3f" %(results.mean()))
```

```python
kfold = KFold(n_splits =10 , shuffle=True, random_state=0)
clf = RandomForestClassifier()
scoring = 'accuracy'
results = cross_val_score(clf, x_train, y_train, cv= kfold, scoring=scoring)
print(results)
print("10 Fold cross validation average accuracy: %.3f" %(results.mean()))
```

```python
# Random Forest and Logistic seems good model. But Decision tree is overfitting
# Based on accuracy using kfolds. I choose random forest as the best model. So I will do hyperparameter tuning on it
```

There are 2 false in this case:

    False Negative = actual Cancel , prediction Not cancel
    
    False Positive = actual Not cancel, prediction Cancel

For me, we have to minimize False Positive. Because if that costumer actually come to hotel. But the hotel predict not. I will cause a lot of mess. Because, hotel don't make a call before to confirm about costumer arrival time, check in time, and many more. So maybe, the room is not ready yet or the customer don't get a special service from hotel like airport pick up or welcome drink & food. The worst thing that may happen is the room already given to another customer. There is no more room available

For False Negative, I think it's okay. Because hotel always do confirm before. So when they make a call to costumer. They may know that customer actually cancel the booking. NO LOSS! NO DISADVANTAGE!

Because I want to minimize False Positive, so I choose model with high precision

```python
models = [LogisticRegression(),KNeighborsClassifier(),DecisionTreeClassifier(),
          RandomForestClassifier()]
names = ["LogisticRegression","KNN","DecisionTree","Random_Forest"]
```

```python
from sklearn.metrics import precision_score
for name, model in zip(names, models):
    model.fit(x_train, y_train)
    y_pred = model.predict(x_test) 
    print(name,':',"%.3f" % (precision_score(y_test, y_pred)))
```

It's good. There are 2 models with high precision ( DECSION TREE OVERFIT *we've checked before)

Let's see False Positive in number

```python
for name, model in zip(names, models):
    model.fit(x_train, y_train)
    y_pred = model.predict(x_test)
    print(name,':' )
    f, ax = plt.subplots(figsize=(8,5))
    sns.heatmap(confusion_matrix(y_test, y_pred), annot=True, fmt=".0f", ax=ax)
    plt.xlabel("y Prediksi")
    plt.ylabel("y Aktual")
    plt.show()
```

From the confusion matrix heatmap above, we could see that Random Forest and Logistic Regression has Zero False Positive. 

But we have to choose best of the best? So, let's go back to the accuracy

Random forest has higher accuracy & the model is not overfitting or underfiting. 

BEST MODEL : RANDOM FOREST (0,995)
